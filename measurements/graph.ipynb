```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
file = open("settings.txt", "r")
m = file.readlines()
diskr = float(m[0][:-1])
u_step = float(m[1])
file.close()
```

```python
U = np.loadtxt("data.txt")
U = U * u_step
all_time = 1200/diskr
T = np.linspace(0, all_time, 1200)
```

```python
fig, ax = plt.subplots(figsize = (8, 6), dpi = 300)
ax.set_title(r"$U(t)$ для конденсатора в RC - цепочке")
ax.minorticks_on()
ax.grid(which = 'major', color = 'black')
ax.grid(which = 'minor')

ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

ax.yaxis.set_major_locator(ticker.MultipleLocator(0.25))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))

ax.set_ylabel("U, B")
ax.set_xlabel("t, с")

ax.text(0, 2,  "Какой-то текст", fontsize = 19)

ax.plot(T[0:600], U[0:600], color = "red", marker = '^', markevery = 30, ms = 5, mec = "blue", mfc = "blue", label = r"$U(t)$")
ax.plot(T[600:900], U[600:900], color = "red", marker = '^', markevery = 10, ms = 5, mec = "blue", mfc = "blue")
ax.plot(T[900:], U[900:], color = "red", marker = '^', markevery = 30, ms = 5, mec = "blue", mfc = "blue")

ax.legend(shadow  = False, loc = 'upper left', fontsize = 10)

plt.savefig("graph.png")
```
